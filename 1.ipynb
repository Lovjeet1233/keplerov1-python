```python
from elevenlabs import ElevenLabs
from elevenlabs import save
from dotenv import load_dotenv

# Initialize ElevenLabs client (correct for your SDK version)
client = ElevenLabs(
    base_url="https://api.eu.residency.elevenlabs.io",
    api_key="sk_3e125ea53a3d12cceba92259a3b2e591b7b98671957e9d30_residency_eu"
)

def text_to_audio(text, file_path="output.mp3", voice_id="QABTI1ryPrQsJUflbKB7"):
    """
    Convert text to speech using ElevenLabs and save as audio file.
    """
    audio = client.text_to_speech.convert(
        voice_id=voice_id,
        model_id="eleven_flash_v2_5",
        text=text
    )

    save(audio, file_path)
    print(f"Audio saved successfully: {file_path}")


if __name__ == "__main__":
    text = "Hello, how can I help you?"
    text_to_audio(text)
```
